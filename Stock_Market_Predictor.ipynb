```python
#For Backtesting
def predict(train, test,predictors,model):
    model.fit(train[predictors], train["Target"])# model learned the patterns from training data
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
def backtest(data,model, predictors, start=2500, step =250):#stock market open for 250 days and we have to do 10 years
    all_predictions = []

    for  i in range(start, data.shape[0], step):
      train = data.iloc[0:i]. copy()
      test = data.iloc[i:(i+step)].copy()
      predictions = predict(train, test, predictors, model)
      all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
class LogisticRegressionGD:
    def __init__(self, learning_rate=0.01, n_iterations=2000, verbose=True):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.verbose = verbose
        self.weights = None
        self.bias = None
        self.costs = []

    def _sigmoid(self, z):
        """The sigmoid activation function."""
        # Clip z to avoid overflow in np.exp for large negative values
        z = np.clip(z, -500, 500)
        return 1 / (1 + np.exp(-z))

    def fit(self, X, y):
        n_samples, n_features = X.shape

        # Initialize weights and bias
        self.weights = np.zeros(n_features)
        self.bias = 0

        for i in range(self.n_iterations):
            # Calculate the linear model (z = w.x + b)
            linear_model = np.dot(X, self.weights) + self.bias
            # Apply the sigmoid function to get predictions (probabilities)
            y_predicted = self._sigmoid(linear_model)

            # --- Calculate Binary Cross-Entropy Cost ---
            # We add a small epsilon value to avoid log(0) errors
            epsilon = 1e-15
            cost = - (1 / n_samples) * np.sum(y * np.log(y_predicted + epsilon) + (1 - y) * np.log(1 - y_predicted + epsilon))
            self.costs.append(cost)

            if self.verbose and i % 100 == 0:
                print(f"Iteration {i}: Cost = {cost:.4f}")

            # --- Compute Gradients (derivatives of the cost function) ---
            dw = (1 / n_samples) * np.dot(X.T, (y_predicted - y))
            db = (1 / n_samples) * np.sum(y_predicted - y)

            # --- Update Weights and Bias using Gradient Descent ---
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

    def predict_proba(self, X):
        """Return probability estimates for the positive class (1)."""
        linear_model = np.dot(X, self.weights) + self.bias
        return self._sigmoid(linear_model)

    def predict(self, X, threshold=0.5):
        """
        Return class labels for samples in X based on a probability threshold.
        """
        probabilities = self.predict_proba(X)
        return [1 if i > threshold else 0 for i in probabilities]

print("Model class created.")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import matplotlib.pyplot as plt
import seaborn as sns
print("Step 1: Preparing data for the LSTM model...")

# --- Feature Scaling ---
# Neural networks perform best when input features are scaled to a 0-1 range.
scaler = MinMaxScaler(feature_range=(0, 1))

# Combine all predictors you've created into one list
all_predictors = new_predictors + ["Close", "Volume", "Open", "High", "Low"]
scaled_data = scaler.fit_transform(sp500[all_predictors])
target_data = sp500["Target"].values
```

```python
#creating sequences of the last 60 days of data to predict the next day.
def create_sequences(data, target, time_steps=60):
    X, y = [], []
    for i in range(time_steps, len(data)):
        X.append(data[i-time_steps:i])
        y.append(target[i])
    return np.array(X), np.array(y)

TIME_STEPS = 60
X, y = create_sequences(scaled_data, target_data, TIME_STEPS)

# --- Train/Test Split ---.
split_index = int(len(X) * 0.8)
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]

print(f"Data prepared with {X_train.shape[0]} training samples and {X_test.shape[0]} testing samples.")
print(f"Shape of X_train (samples, timesteps, features): {X_train.shape}")
```

```python
#STEP 2: BUILD AND TRAIN THE LSTM MODEL
# --------------------------------------------------------------------------
print("\nStep 2: Building and training the LSTM model...")

# Defining the model architecture
model_lstm = Sequential([
    LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dropout(0.2),
    LSTM(units=50, return_sequences=False),
    Dropout(0.2),
    Dense(units=25),
    Dense(units=1, activation='sigmoid') # Sigmoid for binary classification (outputs a probability)
])
```

```python
# Compile the model
model_lstm.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model_lstm.summary()
```

```python
history = model_lstm.fit(X_train, y_train, epochs=25, batch_size=32, validation_split=0.1, verbose=1)

print("\nModel training complete.")
```

```python
print("\nStep 3: Evaluating the LSTM model...")

# Make predictions (these are probabilities)
predictions_proba = model_lstm.predict(X_test)

# Convert probabilities to class labels (0 or 1) using a 0.5 threshold
predictions = (predictions_proba > 0.5).astype(int)

# --- Calculate Precision ---
precision = precision_score(y_test, predictions)
print(f"\nPrecision on the test set: {precision:.4f}")
```

```python
plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss During Training')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

**Objective**
To develop and evaluate machine learning models for predicting the daily price direction of the S&P 500 index. The task was treated as a binary classification problem to predict whether the next day's closing price would be higher (1) or not (0).

**Methodology**
The project involved several key stages:

**Data Acquisition**: Historical data for ^GSPC was sourced via the yfinance API.

**Feature Engineering**: Predictors were created using rolling averages and trend analysis over multiple time horizons to capture market momentum.

**Backtesting**: A robust backtesting function was implemented to simulate historical performance and provide a realistic evaluation metric.

**Model Implementation**: Two models were developed and compared:

A RandomForestClassifier with a tuned probability threshold.

A Long Short-Term Memory (LSTM) neural network, designed for sequential data.

**Results**
The performance of both models was evaluated using the precision score, which measures the accuracy of positive predictions (i.e., "up" days).

**Conclusion**

The analysis confirms that with appropriate feature engineering and model selection, it is possible to achieve a predictive precision that surpasses a random baseline. The LSTM network is the superior model for this specific time-series forecasting task.
